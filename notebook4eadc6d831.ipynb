```python
import pandas as pd
import numpy as np

file_path = "/kaggle/input/datasets/amar5693/student-performance-dataset/ultimate_student_productivity_dataset_5000.csv"
df = pd.read_csv(file_path)

numeric_df = df.select_dtypes(include=[np.number])

selected_columns = numeric_df.columns[:5]

small_df = numeric_df[selected_columns].head(10)

print("===== SELECTED DATA (First 10 Rows, 5 Columns) =====")
print(small_df)

stats_table = pd.DataFrame({
    "Mean": small_df.mean(),
    "Median": small_df.median(),
    "Mode": small_df.mode().iloc[0],
    "Variance": small_df.var(),
    "Std_Deviation": small_df.std()
})

print("\n===== STATISTICAL MEASURES =====")
print(stats_table)
```
